# Résultat : population sous l'aire

``````python
import shapefile
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon

import geopandas as gpd

file_path = "/content/gdrive/MyDrive/Cartographie/Données sites/agriviva.kml"
``````

``````python
!pip install fastkml
``````

``````python
from fastkml import kml

with open(file_path,'rb') as myfile:
   doc = myfile.read()
k = kml.KML()
k.from_string(doc)
``````

``````python
from lxml import etree

data = open(file_path,'rb')
xml_content = data.read()
root = etree.XML(xml_content)
``````

``````python
# Find the coordinates within any Polygon element
polygon_coordinates_elements = root.xpath('//Polygon//coordinates')

if polygon_coordinates_elements:
    # Extract the coordinates from the first found Polygon element
    polygon_coordinates = polygon_coordinates_elements[0].text.strip()

    # Split and convert coordinates to a list of tuples
    coordinates_list = [tuple(map(float, coord.split(','))) for coord in polygon_coordinates.split()]

    # Print the extracted coordinates
    print("Polygon Coordinates:")
    for coord in coordinates_list:
        print(coord)
else:
    print("No <coordinates> element found within any Polygon element in the KML file.")
``````
